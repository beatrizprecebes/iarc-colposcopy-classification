```python
def train_and_validation_step(self, epochs):
        optimizer = torch.optim.AdamW(params=self.model.parameters(), lr=self.lr)
        early_stopper = EarlyStopper(patience=5, min_delta=10)
        
        total_training_loss = list()
        total_validation_loss = list()

        metric_fb = []
        metric_cohen_kappa = []
        metric_acc_balanced = []

        best_loss = 1000

        for epoch in range(1, epochs+1):
            # train_loss = 0
            valid_loss = 0
            y_pred = []
            y_true = []
        
            print(f"Epoch: {epoch}/{epochs}")
        
            for i, batch in enumerate(tqdm(colpo_train_dataloader)):
                self.model.train()
        
                inputs, labels = batch['pixel_values'], batch['labels']
                outputs = self.model(pixel_values=inputs, labels=labels)
        
                loss = outputs[0]
                loss.mean().backward(retain_graph=True)
                train_loss += loss.mean().item()
        
                optimizer.step()
                optimizer.zero_grad()
        
            total_training_loss.append(train_loss/len(colpo_train_dataloader))

            for j, batch in enumerate(tqdm(colpo_val_dataloader)):
                self.model.eval()
        
                inputs, labels = batch['pixel_values'], batch['labels']
                outputs = self.model(pixel_values=inputs, labels=labels)
        
                loss = outputs[0]
                loss.mean().backward(retain_graph=True)
                valid_loss += loss.mean().item()
        
                logits = outputs.logits
        
                y_pred = np.concatenate([y_pred, torch.argmax(logits, dim=-1).tolist()])
                y_true = np.concatenate([y_true, labels])
        
            total_validation_loss.append(valid_loss/len(colpo_val_dataloader))

            valid_fb = compute_metrics(y_true, y_pred, 'fb_score')
            valid_kappa = compute_metrics(y_true, y_pred, 'cohen_kappa')
            valid_acc_b = compute_metrics(y_true, y_pred, 'balanced_accuracy')

            metric_fb.append(valid_fb)
            metric_cohen_kappa.append(valid_kappa)
            metric_acc_balanced.append(valid_acc_b)
        
            print(f"train_loss: {round(train_loss/len(colpo_train_dataloader), 3)}",
                  f"- val_loss: {round(valid_loss/len(colpo_val_dataloader), 3)}",
                  f"- fb_score: {round(valid_fb, 3)}",
                  f"- cohen_kappa: {round(valid_kappa, 3)}",
                  f"- balanced acc: {round(valid_acc_b, 3)}")
        
            if(valid_loss/len(colpo_val_dataloader) < best_loss):
                best_loss = valid_loss/len(colpo_val_dataloader)
                torch.save({
                    'epoch': epoch,
                    'model_state_dict': self.model.state_dict()
                }, basepath_da_e+"/binary_vit_rd_da_equalizer.pth")
        
            if early_stopper.early_stop(valid_loss):
                break
```
